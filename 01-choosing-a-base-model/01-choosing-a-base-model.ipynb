## What Makes a Good Base Model?

A good base model matches your task, fits your hardware, and has license terms you can live with.

Important factors:

* Size: smaller models are easier to run and debug.
* Task alignment: chat tuned models are easier for instruction following.
* License: some models are very open, some are restricted, and some require access approval.
* Community support: popular models usually have more examples and troubleshooting posts.

## Parameter Count Tradeoff

A rough beginner rule:

* 1B class models are the easiest place to learn.
* 3B class models often give a better quality to cost balance.
* 7B class models are a common practical target for real work.
* 13B and above usually move you into stronger GPU requirements.

```python
hardware_view = [
    {"parameter_count": "1B", "cpu_demo": "Reasonable", "free_colab_gpu": "Comfortable", "notes": "Best for learning loops and debugging."},
    {"parameter_count": "3B", "cpu_demo": "Slow", "free_colab_gpu": "Usually workable", "notes": "Good step up once the basics make sense."},
    {"parameter_count": "7B", "cpu_demo": "Impractical", "free_colab_gpu": "Needs LoRA or QLoRA", "notes": "Common real world size for adapter training."},
    {"parameter_count": "13B", "cpu_demo": "Not practical", "free_colab_gpu": "Usually too large for free tier", "notes": "Better for paid GPUs or stronger local rigs."},
]
for row in hardware_view:
    print(row)
```

## Browsing Hugging Face Hub

The Hugging Face Hub lets you filter models by task, license, downloads, and tags.
That is usually where you start when looking for a candidate.

```python
from huggingface_hub import HfApi

api = HfApi()
candidates = list(api.list_models(search="tinyllama", limit=5))
[{"id": model.id, "downloads": model.downloads, "likes": model.likes} for model in candidates]
```

## Decision Matrix

| Use case | Good starting family | Why |
| --- | --- | --- |
| First fine tune on free hardware | TinyLlama or Phi | Small enough to debug without huge cost |
| General instruction following | Mistral or LLaMA | Strong chat performance and broad examples |
| Tight license preference | Phi or Apache licensed Mistral models | Fewer access and reuse surprises |
| Larger consumer GPU project | Qwen or 7B class Mistral | More capacity once you outgrow 1B models |

```python
recommended_models = [
    "microsoft/phi-2",
    "TinyLlama/TinyLlama-1.1B-Chat-v1.0",
]
print("Recommended for later notebooks:")
for model_name in recommended_models:
    print(f"  * {model_name}")
```

## Key Takeaways

* Start with a model that fits your hardware.
* License checks are part of model selection, not an afterthought.
* Small models are better for learning the workflow.
* For this repo, TinyLlama and Phi 2 are the most practical starter choices.
